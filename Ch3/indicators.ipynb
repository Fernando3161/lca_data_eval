```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

cm = 1/2.54

data_adp = "ADP_expanded.xlsx"
df_adp = pd.read_excel(data_adp, engine="openpyxl", sheet_name="ADP-Elements")
df_adp= df_adp[['Symbol','ADP', 'SR','EI','GPR']]
df_adp.columns= ['Symbol','ADP', 'CWADP-SR','CWADP-EI','GPSR']
df_adp = df_adp.dropna()

# divide by the highes gpr
gpr_max=max(df_adp["GPSR"])
df_adp["GPSR"] *= 1/gpr_max 
mat_list= ["Ta", "Mo", "Li", "Gd", "Cu", "Co", "Ce", "Al","Au", "Pt", "Ag"]

df_adp = df_adp[(df_adp["Symbol"].isin(mat_list))]
 
#df_adp.set_index("Symbol", inplace=True)
#df_adp
```

```python
import matplotlib.pyplot as plt 
sns.set_theme()
sns.set(context="paper", style="whitegrid", font="Times New Roman", font_scale = 3)
#sns.set_palette('husl')
sns.set_palette('Greys')
fig, ax = plt.subplots(figsize=(16,10))

df_adp.plot.bar(ax=ax)

ax.set_xlabel("Material")
ax.set_ylabel("Indicator Value")
ax.legend(loc="upper left", bbox_to_anchor=(1,1), title = "Indicator", fontsize='small')
ax.set_xticklabels(mat_list)

plt.yscale("log")
plt.tight_layout()    

plt.savefig("indicators_v2_greys.png", dpi=600)
```
