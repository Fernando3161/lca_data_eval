```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("data_centers.csv", index_col="Data Center Type")
df
```

```python
palette_color = sns.color_palette('dark')
import matplotlib.pyplot as plt

plot = df.transpose().plot.pie(subplots=True, figsize=(11,6))
sns.set_style()
#plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
sns.set(font_scale=1.5,style = "whitegrid", font = "Times New Roman")

fig,axs = plt.subplots(1,2, figsize=(12,6))

df2=df.transpose()
colors = sns.color_palette("viridis")[0:5]
patches = axs[0].pie(df2["Servers"], colors = colors, autopct='%.0f%%')
labels = [x for x in df.columns]
axs[0].legend(labels, loc = "best",title="técnica")

axs[1].pie(df2["Data Center"], colors = colors, autopct='%.00f%%')

plt.show()
```
